```python
import numpy as np
import random
from matplotlib.animation import FuncAnimation
from matplotlib import colors
```

```python
def findOccurrences(s, ch):
    return [i for i, letter in enumerate(s) if letter == ch]


```

```python
def Text_to_List(textfile):
    new_list= []
    with open(textfile) as f:
        text = f.read()
        list_of_text = text.split('\n')
        for word in list_of_text:
            new_word = word.strip(',./""'':;?!')
            new_list.append(new_word)
    return(list_of_text)
```

```python
def Text_to_List2(textfile):
    new_list= []
    with open(textfile) as f:
        text = f.read()
        list_of_text = text.split(' ')
        for word in list_of_text:
            new_word = word.strip(',./""'':;?!')
            new_list.append(new_word)
    return(list_of_text)
```

```python
def Hidden_word(textfile,difficulty):
    Dictionary1 = Text_to_List(textfile)
    Easy_Dictionary = []
    Medium_Dictionary = []
    Expert_Dictionary = Text_to_List2("Tough.txt")
    for word in Dictionary1:
        if len(word) <= 6:
            Easy_Dictionary.append(word.upper())
        else:
            Medium_Dictionary.append(word.upper())
            
    if difficulty.upper() == "EASY":
        Hidden_word = random.choice(Easy_Dictionary)
        return(Hidden_word)
    elif difficulty.upper() == "MEDIUM":
        Hidden_word = random.choice(Medium_Dictionary)
        return(Hidden_word)
    elif difficulty.upper() == "EXPERT":
        Hidden_word = random.choice(Expert_Dictionary)
        return(Hidden_word.upper())
    
```
